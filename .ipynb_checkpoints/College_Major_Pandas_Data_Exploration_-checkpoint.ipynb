```python
# import pandas
import pandas as pd
```

```python
# read the csv file
df = pd.read_csv('salaries-by-college-major.csv')
```

```python
# look at the python dataframe we've just created with .head()
# this will show us the first 5 rows of our dataframe
df.head()
```

```python
# see the number of rows and columns using the "shape" attribute
df.shape
```

```python
# access the column names directly with the "columns" attribute
df.columns
```

```python
# try and figure out if there are any missing or junk data in our dataframe
# NaN(Not a Number) are blank cells or cells that contain strings instead of numbers
# use the ".isna()" method to identify NaN
df.isna()
```

```python
# check the last couple of rows in the dataframe with ".tail()"
df.tail()
```

```python
# delete the last row either manually or simply use the ".dropna()" method from pandas
# create a new dataframe without the last row and examine the last 5 rows to make sure the last row is removed
clean_df = df.dropna()
clean_df.tail()
```

```python
# to access a particular column from a dataframe we can use the square bracket notation like so
clean_df['Starting Median Salary']
```

```python
# to find the highest salary simply chain the ".max()" method
clean_df["Starting Median Salary"].max()
```

```python
# the ".idxmax()" method give us index for the row with the largest value
clean_df["Starting Median Salary"].idxmax()
```

```python
# to see the name of the major that corresponds to that particular row, we use the ".loc" location property as
clean_df["Undergraduate Major"].loc[43]

# here we are selecting both a column(Undergraduate Major) and a row at index 43, so we are retreiving the value of a particular cell
```

```python
# can use the double square brackets notation to achieve exactly the same thing as above
clean_df["Undergraduate Major"][43]
```

```python
# if we don't specify a particluar column we can use the .loc property to retrieve an entire row
clean_df.loc[43]
```

```python
# if we have multiple lines in the same cell, only the last one will get printed as an output automatically
# if we like to see more than one thing printed out use a print statement
# the highest mid-career salary
print(clean_df["Mid-Career Median Salary"].max())
print(f"Index for the max mid career salary:{clean_df['Mid-Career Median Salary'].idxmax()}")
clean_df["Undergraduate Major"][8]
```

```python
# the lowest starting and mid-career salary
print(clean_df['Starting Median Salary'].min())

# nested the code as seen in the previous lessions in same line
clean_df["Undergraduate Major"].loc[clean_df['Starting Median Salary'].idxmin()]
```

```python
# access the row at the index of the smallest mid-career salary as
clean_df.loc[clean_df['Mid-Career Median Salary'].idxmin()]
```

```python
# lowest risk majors
# pandas allows us to do simple arithmetic with entire columns
# to find difference between two columns use:
clean_df["Mid-Career 90th Percentile Salary"]-clean_df["Mid-Career 10th Percentile Salary"]

# or use the ".subtract()" method
clean_df["Mid-Career 90th Percentile Salary"].subtract(clean_df["Mid-Career 10th Percentile Salary"])
```

```python
# the output of the above computation will be another Pandas dataframe column
# add this column to our existing dataframe with the ".insert()" method
# arguments are .insert(position of where column should be inserted, Name of column, pandas dataframe)
spread_col = clean_df["Mid-Career 90th Percentile Salary"]-clean_df["Mid-Career 10th Percentile Salary"]
clean_df.insert(1, 'Spread', spread_col)
clean_df.head()
```

```python
# sorting by lowest spread
# to see which degrees has the lowest spread, we can use the ".sort_values" method
low_risk = clean_df.sort_values('Spread')

# to see only the name of the degree and spread value, we can pass a list of these two column names to look at the .head() of these two columns exclusively
low_risk[['Undergraduate Major','Spread']].head()
```

```python
# Majors with the highest potential
highest_potential = clean_df.sort_values('Mid-Career 90th Percentile Salary', ascending=False)
highest_potential[['Undergraduate Major', 'Mid-Career 90th Percentile Salary']].head()
```

```python
# Majors with the Greatest spread in salaries
highest_spread = clean_df.sort_values('Spread', ascending=False)
highest_spread[['Undergraduate Major','Spread']].head()
```

```python
highest_spread = clean_df.sort_values('Mid-Career Median Salary', ascending=False)
highest_spread[['Undergraduate Major', 'Mid-Career Median Salary']].head()
```

```python
# Finding which category of degrees(STEM, Business or HASS) has the highest average salary using ".groupby()" method
# count how many majors we have in each category
clean_df.groupby('Group').count()
```

```python
# use the ".mean()" method to find the average salary by group
clean_df.groupby('Group').mean()
```

```python
# number format in the output
# can tell pandas to print the numbers in our notebook to look like 1,0123.23 with the following line:
pd.options.display.float_format = '{:,.2f}'.format
clean_df.groupby('Group').mean()
```
